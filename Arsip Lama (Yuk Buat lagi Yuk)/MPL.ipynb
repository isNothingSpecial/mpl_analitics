```python
import pandas as pd
```

```python
df = pd.read_csv("Match MPL.csv")
df
```

```python
df.columns
```

```python
df1 = pd.DataFrame(df, columns=["Match ID", "Blue Team","Blue Team Heroes","Red Team","Red Team Heroes", "Winner"])
df1
```

```python
df2 = pd.DataFrame(df1)
df2.rename(columns={'Blue Team Heroes': 'Blue Team EXP Laner'}, inplace=True)
df2.insert(3, 'Blue Team Jungler', 0)
df2.insert(4, 'Blue Team Midlaner', 0)
df2.insert(5, 'Blue Team Gold Laner', 0)
df2.insert(6, 'Blue Team Roamer', 0)
df2
```

```python
#loop for each row, adding heroes to new columns

df3 = pd.DataFrame(df2)

#change result column to 1 and 0. 1 means winning, 0 means losing
for i in range(len(df3)): 
    if (df3.loc[i,'Winner'] == "Blue Team"):
        df3.loc[i,'Winner'] = 1
    elif (df3.loc[i,'Winner'] == "Red Team"):
        df3.loc[i,'Winner'] = 0
        
    data=df3.loc[i,'Blue Team EXP Laner']
    data=data.split(",")
    
    df3.loc[i,'Blue Team EXP Laner']=data[0].strip()
    df3.loc[i,'Blue Team Jungler']=data[1].strip()
    df3.loc[i,'Blue Team Midlaner']=data[2].strip()
    df3.loc[i,'Blue Team Gold Laner']=data[3].strip()
    df3.loc[i,'Blue Team Roamer']=data[4].strip()
    
df3
```

```python
df4 = pd.DataFrame(df3)
df4.rename(columns={'Red Team Heroes': 'Red Team EXP Laner'}, inplace=True)
df4.insert(9, 'Red Team Jungler', 0)
df4.insert(10, 'Red Team Midlaner', 0)
df4.insert(11, 'Red Team Gold Laner', 0)
df4.insert(12, 'Red Team Roamer', 0)
df4
```

```python
#loop for each row, adding heroes to new columns

df5 = pd.DataFrame(df4)

#change result column to 1 and 0. 1 means winning, 0 means losing
for i in range(len(df5)): 
    if (df5.loc[i,'Winner'] == 1):
        df5.loc[i,'Winner'] = 1
    elif (df5.loc[i,'Winner'] == 0):
        df5.loc[i,'Winner'] = 0
        
    data1=df5.loc[i,'Red Team EXP Laner']
    data1=data1.split(",")
    
    df5.loc[i,'Red Team EXP Laner']=data1[0].strip()
    df5.loc[i,'Red Team Jungler']=data1[1].strip()
    df5.loc[i,'Red Team Midlaner']=data1[2].strip()
    df5.loc[i,'Red Team Gold Laner']=data1[3].strip()
    df5.loc[i,'Red Team Roamer']=data1[4].strip()
    
df5
```
